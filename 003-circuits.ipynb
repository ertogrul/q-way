```python
from qiskit import *
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
# create a quantum circiut with one q bit and one classical bit
circuit = QuantumCircuit(1,1)
```

```python
# that circuit will have one operation on it
circuit.x(0)
# lets now take that circuit and simulate it on state vector simulator
# (state vector is a vector that describes a state of qbit)
simulator = Aer.get_backend('statevector_simulator')
# lets execute this circuit on the backend
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
print(statevector)

# result:
# vector 0 with 0 imaginary component and vector 1 with 0 imaginary component
```

```python
%matplotlib inline
circuit.draw(output='mpl')
```

```python
plot_bloch_multivector(statevector)
# quantum stated for individual qbit can be represented on the sphere
# operations on the states is often called 'rotations on the sphere'
# what the x gate done is take us from the state 0 to state 1
# state one is the bottom of the sphere
```

```python
# measure qbit 0 and put that result into classical bit 09
circuit.measure([0], [0])
# set up the backend
backend = Aer.get_backend('qasm_simulator')
# execute this on the backend and do 1024 shots
result = execute(circuit, backend = backend, shots = 1024).result()
# get counts
counts = result.get_counts()
# plot results
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
# to make a matrix representation we just need to change
# backend to unitary_simulator
circuit = QuantumCircuit(1,1)
circuit.x(0)
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend = simulator).result()
unitary = result.get_unitary()
print(unitary)
```
